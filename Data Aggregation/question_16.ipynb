How do you group data by a column and calculate summary statistics?
Dataset: Iris dataset.

```python
import pandas as pd

df = pd.read_csv('iris.data.csv')
print(df.info())
```

```python
summary = df.groupby('5.1').agg({
    '0.2' : ['mean', 'std', 'min', 'count', 'max']
})

# rename group for clarity
summary.columns = ['mean', 'std', 'min', 'count', 'max']
```

```python
print(summary)
```
