How do you pivot a DataFrame?
Dataset: Sales data with product categories.

```python
import pandas as pd

# Sample sales data
data = {
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02', '2022-01-03'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Category': ['X', 'Y', 'X', 'Y', 'X'],
    'Sales': [100, 200, 150, 250, 120]
}

# Create DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame
pivot_df = df.pivot_table(index='Date', columns=['Category', 'Product'], values='Sales', aggfunc='sum')

# Reset index for clarity
pivot_df.reset_index(inplace=True)

# Display pivoted DataFrame
print("Pivoted DataFrame:")
print(pivot_df)
```
