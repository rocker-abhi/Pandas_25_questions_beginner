# question 18 :
How do you calculate the average sales per month?
Dataset: Sales data with dates.


```python
import pandas as pd

data = {
    'Date': ['2022-01-01', '2022-01-01', '2022-01-02', '2022-01-02', '2022-01-03'],
    'Product': ['A', 'B', 'A', 'B', 'A'],
    'Category': ['X', 'Y', 'X', 'Y', 'X'],
    'Sales': [100, 200, 150, 250, 120]
}

df = pd.DataFrame(data)
print(df)
```

```python
# calculate the average sales per month

pivot_df = df.pivot_table(index='Date', columns=['Product','Category'], values='Sales', aggfunc='sum')

pivot_df.reset_index(inplace=True)

print(pivot_df)
```
