How do you remove duplicate rows from a DataFrame?
Dataset: Wine Quality dataset.


```python
import pandas as pd 

df = pd.read_csv('iris.data.csv')
print(df.info())
```

```python
# find the duplicate in the dataframe
if df.duplicated().any() :
    print("Duplicate found in the dataset .")
else:
    print("Duplicate not found in the dataset .")
```

```python
df = df.drop_duplicates()
# again checking weather the duplicate exist in the database .
if df.duplicated().any():
    print('duplicate found')
else:
    print('duplicate not found in the dataframe')
```
