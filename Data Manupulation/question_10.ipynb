How do you apply a function to each element of a DataFrame?
Dataset: Wine Quality dataset.

```python
import pandas as pd

df = pd.read_csv('WineQT.csv')
print(df)
```

```python
# creating the new dataframe 
df_2 = df
```

```python
# defining function top apply in the dataframe

def my_function(x):
    return x * 2

df['quality'] = df['quality'].apply(my_function)
print(df)
```

```python
# apply the function to the whole dataframe .

df_2 = df_2.apply(my_function)
print(df_2)
```
