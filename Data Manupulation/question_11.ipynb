How do you extract the year from a datetime column in a DataFrame?
Dataset: Netflix stock price dataset.

```python
import pandas as pd
df = pd.read_csv('NFLX.csv')
print(df.info())
```

```python
# convert the date column to datetime field .
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'].dt.year)
```
