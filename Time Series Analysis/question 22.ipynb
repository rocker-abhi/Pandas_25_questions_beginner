How do you resample time series data to a different frequency?

```python
import pandas as pd

# Sample time series data
data = {
    'Date': pd.date_range(start='2022-01-01', end='2022-12-31', freq='D'),
    'Value': range(365)
}

# Create DataFrame with DateTime index
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

print(df)
```

```python

# Resample time series data from daily to monthly frequency
resampled_df = df.resample('ME').sum()  # Resample and sum the values within each month

# Display resampled DataFrame
print("Resampled DataFrame:")
print(resampled_df)
```
