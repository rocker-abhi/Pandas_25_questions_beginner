How do you calculate rolling mean of a time series?

```python
import pandas as pd

# Sample time series data
data = {
    'Date': pd.date_range(start='2022-01-01', end='2022-01-31', freq='D'),
    'Value': range(31)
}

# Create DataFrame with DateTime index
df = pd.DataFrame(data)
df.set_index('Date', inplace=True)

# Calculate rolling mean over a window of 7 days
rolling_mean = df.rolling(window=7).mean()

# Display DataFrame with rolling mean
print("DataFrame with Rolling Mean:")
print(rolling_mean)
```
