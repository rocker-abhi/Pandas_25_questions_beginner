How do you extract day of the week from a datetime column?

```python
import pandas as pd

# Sample data with datetime column
data = {
    'Date': pd.date_range(start='2022-01-01', end='2022-01-07')
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract day of the week from the datetime column
df['Day_of_Week'] = df['Date'].dt.dayofweek

# Display DataFrame with day of the week
print("DataFrame with Day of the Week:")
print(df)
```
