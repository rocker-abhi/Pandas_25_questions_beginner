How do you create a scatter plot between two numerical columns?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 20, 30, 40, 50]
}

# Create DataFrame
df = pd.DataFrame(data)

# Create scatter plot between columns 'X' and 'Y'
df.plot.scatter(x='X', y='Y', color='blue', marker='o', edgecolor='black')

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot between X and Y')

# Show plot
plt.show()
```
