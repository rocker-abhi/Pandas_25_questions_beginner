How do you select rows and columns by index labels?
Dataset: Iris dataset.

```python
import pandas as pd 

df = pd.read_csv('iris.data.csv')
print(df.info())
```

```python
# setting the index labels 
df.index = range(100, 249)

# display original frame
print(df)
```

```python
selected_data = df.loc[[149,161],['5.1','3.5']]
print(selected_data)
```
