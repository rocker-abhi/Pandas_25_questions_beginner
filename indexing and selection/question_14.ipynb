# question 14 :
How do you select rows and columns by integer location?
Dataset: Titanic dataset.

```python
import pandas as pd

df = pd.read_csv('tested.csv')
```

```python
# printing zero index
zero_index = df.iloc[0]
print(zero_index)
```

```python
# selecting multiple rows (e.g row 0-4)

cell_0_5 = df.iloc[0:4]
print(cell_0_5)
```

```python
# selecting multiple cells row 0 col 1 
multiple_cels = df.iloc[0,1]
print(multiple_cels)
```

```python
# selecting the row 0-5 and col 0-3

new_dataset = df.iloc[0:5, 0:3]
print(new_dataset)
```
