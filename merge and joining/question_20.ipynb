How do you perform an outer join between two DataFrames?

```python
import pandas as pd

data1 = {
    'common_column': [1, 2, 3, 4],
    'data1_col': ['A', 'B', 'C', 'D']
}
# Sample data for df2
data2 = {
    'common_column': [2, 3, 4, 5],
    'data2_col': ['X', 'Y', 'Z', 'W']
}

# Create DataFrame df1
df1 = pd.DataFrame(data1)

# Create DataFrame df2
df2 = pd.DataFrame(data2)

# Perform an outer join between DataFrames df1 and df2 by common column
merged_df = pd.merge(df1, df2, on='common_column', how='outer')

# Display merged DataFrame
print("Outer joined DataFrame:")
print(merged_df)
```
