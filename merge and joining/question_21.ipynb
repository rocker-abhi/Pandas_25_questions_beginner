How do you concatenate multiple DataFrames vertically?

```python
import pandas as pd

# Sample data for sales data for different years
data_2020 = {
    'Date': ['2020-01-01', '2020-01-02', '2020-01-03'],
    'Sales': [100, 200, 150]
}

data_2021 = {
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03'],
    'Sales': [120, 180, 220]
}

data_2022 = {
    'Date': ['2022-01-01', '2022-01-02', '2022-01-03'],
    'Sales': [150, 250, 200]
}

# Create DataFrames for sales data for different years
df_2020 = pd.DataFrame(data_2020)
df_2021 = pd.DataFrame(data_2021)
df_2022 = pd.DataFrame(data_2022)

# Concatenate DataFrames vertically
concatenated_df = pd.concat([df_2020, df_2021, df_2022], axis=0)

# Reset index for clarity
concatenated_df.reset_index(drop=True, inplace=True)

# Display concatenated DataFrame
print("Concatenated DataFrame:")
print(concatenated_df)
```
